```python
import numpy as np
# Constants
m = 1.7
n = 1

# Function to calculate the equilibrium cost based on c, theta, m, and n for the four cases
def calculate_equilibrium_cost(c, theta, m, n):
    if c + theta / 2 > (m + n - 1) / 2:
        equilibrium_cost = (1 - m + n + theta) / 2 + c + m
    elif c <= c + theta / 2 > (m + n - 1) / 2 and (m + n - theta - 2) / 2 < c < (m + n - 1 - theta) / 2:
        equilibrium_cost = m + n
    elif n > 1 + c+theta:
        equilibrium_cost = 1 + m + c
    elif m >= 1 + c + theta:
        equilibrium_cost = 1 + n + c + theta
    else:
        equilibrium_cost = None
    return equilibrium_cost

c_values = np.linspace(0, 1, 10)  # 10 evenly spaced values from 0 to 1
theta_values = np.random.uniform(0, 1, 100)  

all_results = []
for c in c_values:
    for theta in theta_values:
        cost = calculate_equilibrium_cost(c, theta, m, n)
        all_results.append((c, theta, cost))

# Filter the results to find those cases where the equilibrium cost is less than m + n
results_less_than_m_plus_n = [result for result in all_results if result[2] is not None and result[2] < calculate_equilibrium_cost(c, 0.5, m, n)]

results_less_than_m_plus_n
```
